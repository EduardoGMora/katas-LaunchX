## Ejercicio: Usar listas para almacenar nombres de planetas
En primer lugar, crea una variable denominada planets. Agrega los ocho planetas (sin Plutón) a la lista. A continuación, muestra el número de planetas.

```python
# Creamos la lista planets y la mostramos
planets=['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
print(planets)
numeroplanets=len(planets)
print("numero de planetas sin plutón", numeroplanets);

# Agregamos a plutón y mostramos el último elemento
planets.append('pluton')
print(planets[8])
numeroplanets=len(planets)
print("numero de planetas conn plutón", numeroplanets)


```

# Ejercicio 2: Trabajando con datos de una lista
## Usar slices para recuperar partes de una lista
queremos mostrar planetas más cerca y más lejos del sol de un planeta que el usuario ingresa por teclado.

```python
# Lista de planetas
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Neptune']

# Solicitamos el nombre de un planeta *Pista:  input()*
planeta=str(input("Digita el nombre de un planeta iniciando con mayúsculas "))

# Busca el planeta en la lista
planetabuscado=planets.index(planeta)

# Muestra los planetas más cercanos al sol
print("Aquí estan los planetas más cerca del sol que " + planeta);
print(planets[0:planetabuscado])
# Muestra los planetas más lejanos al sol
print("Acá estan los planetas más lejos del sol que " + planeta);
print(planets[planetabuscado+1:])
```
